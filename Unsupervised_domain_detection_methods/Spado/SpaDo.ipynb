```r
# Lặp từ slice thứ 5 đến 12
for (slice_choose in 5:12) {
  cat("\n=============================\n")
  slice_name <- names(sample_information_decon_list)[slice_choose]
  cat("📌 Slice được chọn là:", slice_name, "\n")
  
  test_coordinate <- test_coordinate_list[[slice_choose]]
  sample_information_decon <- sample_information_decon_list[[slice_choose]]
  sample_information_region <- sample_information_region_list[[slice_choose]]
  
  cell_type_distribution <- SpatialCellTypeDistribution(
    sample_information_coordinate = test_coordinate,
    sequence_resolution = "spot",
    sample_information_decon = sample_information_decon
  )
  
  DD <- DistributionDistance(
    cell_type_distribution,
    distance = "JSD",
    no_cores = 1
  )
  
  library(fastcluster)
  domain_hclust <- DomainHclust(
    distribution_distance = DD,
    autoselection = FALSE,
    domain_num = 7
  )
  
  library(flexclust)
  ARI_cal <- table(
    sample_information_region,
    domain_hclust$hclust_result_df[names(sample_information_region), 7]
  )
  cat("✅ Rand Index: ", randIndex(ARI_cal, correct = TRUE), "\n")
  
#   DomainPlot(
#     domain_hclust = domain_hclust,
#     distribution_distance = DD,
#     sample_information_coordinate = test_coordinate,
#     k = 7
#   )
}
```

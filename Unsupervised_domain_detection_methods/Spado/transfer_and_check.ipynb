```python
import scipy.io
import pandas as pd

for i in range(151508, 151677):
    slice_name = str(i)
    
    try:
        
        adata = sc.read(f"D:/VietBao2025/Methods/Dataset/Maynard/{slice_name}.h5ad")

        # lấy tọa độ spatial
        coords = adata.obsm["spatial"]

        # tạo dataframe
        coords_df = pd.DataFrame(coords, columns=["X", "Y"])

        # thêm cell_id (spot ID)
        coords_df["cell_id"] = adata.obs_names

        # reorder cột
        coords_df = coords_df[["cell_id", "X", "Y"]]

        # lưu file
        coords_df.to_csv(f"{slice_name}_coordinates_h5ad.csv", index=False)
        
    except:
        print(f"❌ Missing slice {slice_name}")
```
